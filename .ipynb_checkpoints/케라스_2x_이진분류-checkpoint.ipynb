```python
import pandas as pd
```

```python
col_names = ['age', 'sex','cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpack', 'slope', 'ca', 'hal', 'heartdisease']
```

```python
len(col_names)
```

```python
data = pd.read_excel('ClevelandData.xlsx', names=col_names)
```

```python
import numpy as np
```

```python
data_new = data.replace('?', np.nan)
```

```python
data_new.isnull().sum()
```

```python
data_new['ca'].value_counts()
```

```python
data_new['hal'].value_counts()
```

```python
data_new_nn = data_new.dropna()
```

```python
data_new_nn.info()
```

```python
data_new_nn.isnull().sum()
```

```python
InputNames = col_names
```

```python
len(InputNames)
```

```python
InputNames.pop()
```

```python
len(InputNames)
```

```python
Input_data = pd.DataFrame(data_new_nn.iloc[:, :13], columns=InputNames)
```

```python
Input_data.head()
```

```python
Input_data.shape
```

```python
Target = pd.DataFrame(data_new_nn.iloc[:, 13], columns=['heartdisease'])
```

```python
Target.head()
```

```python
Target.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(Input_data)
```

```python
InputScaled = scaler.fit_transform(Input_data)
```

```python
type(InputScaled)
```

```python
InputScaled_ = pd.DataFrame(InputScaled, columns=InputNames)
```

```python
InputScaled_.head()
```

```python
InputScaled_.describe().transpose()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(InputScaled_, Target, test_size=0.3, random_state=5)
```

```python
X_train.shape, X_test.shape
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(30, activation='tanh', input_dim=13))
```

```python
model.add(Dense(20, activation='tanh'))
```

```python
model.add(Dense(10, activation='tanh'))
```

```python
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=1000)
```

```python
model.summary()
```

```python
score = model.evaluate(X_test, y_test)
```

```python
score
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, plot_roc_curve
```

```python
confusion_matrix(y_test, y_pred > 0.5)
```

```python
f1_score(y_test, y_pred > 0.5)
```

```python
precision_score(y_test, y_pred > 0.5)
```

```python
recall_score(y_test, y_pred > 0.5)
```
